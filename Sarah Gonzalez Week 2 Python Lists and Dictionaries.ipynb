### **Data Science Skill: List Slicing**

Slicing allows us to "zoom in" on specific parts of our data.

* **Syntax:** `my_list[start:stop:step]`. The index **stop** is **exclusive**, it stops just before this number.
* **The "Exclusive" Rule:** The `stop` index is never included in the result. Think of it as: "Up to, but not including."
* **The "Copy" Rule:** Slicing does not change the original list; it creates a new, smaller version.

**Note:** Slicing is the foundation for data filtering. If you can slice a Python list, you are 50% of the way to understanding how to filter a massive dataset in Pandas!

### **The Difference Between `.sort()` and `sorted()`**

In Python, understanding how to manage data in memory is crucial. The core difference between these two options is **mutability**: one changes the original data (**in-place**), while the other creates a brand new copy (**out-of-place**).

---

#### **1. `.sort()` — The In-Place Modifier**
This is a **method** specifically for lists. It modifies the original list directly and returns `None`.



* **Memory Efficient:** It doesn't create a copy, making it better for massive datasets where RAM is a concern.
* **Destructive:** The original order of the data is lost forever.
* **Restriction:** It only works on **list** objects.

---

**Note:** For Data Science workflows, use `.sort()` only when you are absolutely sure you no longer need the original sequence of your data!

````python
# Original data
numbers = [4, 2, 9, 1]

# Apply the method
numbers.sort()

print(numbers) # Output: [1, 2, 4, 9] (The original list is changed)
````

#### **2. `sorted()` — The Creator**
This is a **built-in function**. It takes an iterable (like a list, tuple, or string) and returns a **new list** containing the sorted elements.



* **Safe:** Your original data remains untouched, allowing you to keep the original sequence for later reference.
* **Flexible:** Unlike the `.sort()` method, this works on **any iterable** (including tuples, dictionaries, and strings).
* **Memory Cost:** It creates a copy, which requires more RAM. This is usually only a factor when working with very large datasets.

---

**Note:** I recommend using `sorted()` as your default. In data science, being able to "audit" your data by comparing the original state to the sorted state is a critical part of the cleaning process.

````python
# Original data
numbers = [4, 2, 9, 1]

# Use the function to create a new variable
new_numbers = sorted(numbers)

print(numbers)     # Output: [4, 2, 9, 1] (Original preserved!)
print(new_numbers) # Output: [1, 2, 4, 9] (The new copy)
````

````python
# TODO
# Case insensitive check by converting both (grep and model) to lowercase

def get_all_matching_models(cars=cars, grep='trail'):
    """return a list of all models containing the case insensitive
    'grep' string which defaults to 'trail' for this exercise,
    sort the resulting sequence alphabetically"""
    results = []
    for models in cars.values():
        for model in models:
            # Case insensitive check by converting both to lowercase
            if grep.lower() in model.lower():
                results.append(model)
    return sorted(results)
````

## While Loop
While loop is used to execute a block of statements repeatedly until a given condition is satisfied. When the condition becomes false, the line immediately after the loop in the program is executed.

![While Loop Flowchart](whileloop.webp)

In below code, loop runs as long as the condition cnt < 3 is true. It increments the counter by 1 on each iteration and prints "Hello Students" three times.

````python
# TODO
cnt = 0
while (cnt < 3):
    cnt = cnt + 1
    print("Hello Student")
````

## Loop control statements
Loop control statements change execution from their normal sequence. When execution leaves a scope, all automatic objects that were created in that scope are destroyed. Python supports the following control statements. 

### Continue Statement
It returns the control to the beginning of the loop. The continue statement is used to skip the current iteration of a loop and move to the next iteration. It is useful when we want to bypass certain conditions without terminating the loop.

````python
for letter in 'pythonisgreat':
    if letter == 't' or letter == 'n':
        continue
    print('Current Letter :', letter)
````

### Break Statement
The break statement in Python brings control out of the loop.

````python
for letter in 'pythonisgreat':
    if letter == 't' or letter == 'n':
        break
    print('Current Letter :', letter)
````

**Explanation**: The loop starts from the first character 'p' and continues until it encounters 't'. When 't' is found, the break statement stops the loop immediately. Since 't' is the value at which the loop stops, it is printed outside the loop

### Pass Statement
We use pass statement in Python to write empty loops. Pass is also used for empty control statements, functions and classes. The loop iterates over each letter in 'pythonisgreat' but doesn't perform any operation, and after the loop finishes, the last letter ('t') is printed.

````python
for letter in 'pythonisgreat':
    pass
print('Last Letter :', letter)
````

### Difference between for loop and while loop in Python

| Feature | For loop | While loop |
| :--- | :--- | :--- |
| **Basic Definition** | Used to iterate over a sequence of items. | Used to repeatedly execute a block of statements while a condition is true. |
| **Design Purpose** | Designed for iterating over a sequence of items (e.g., list, tuple, range). | Used when the number of iterations is not known in advance or to repeat until a condition is met. |
| **Requirement** | Requires a sequence to iterate over. | Requires an initial condition that is tested at the beginning of the loop. |
| **Typical Use Case** | Typically used for iterating over a fixed sequence of items. | Used for more complex control flow situations. |
| **Efficiency** | More efficient when iterating over sequences since the number of iterations is predetermined. | May be more efficient in situations where the condition being tested can be evaluated quickly. |

# List Comprehension
List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

Example:

Based on a list of fruits, you want a new list, containing only the fruits with the letter "a" in the name.

Without list comprehension you will have to write a for statement with a conditional test inside:

````python
# TODO
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)

print(newlist)
````

With list comprehension you can do all that with only one line of code:

````python
# TODO
newlist = [fruit for fruit in fruits if "a" in fruit]
print(newlist)
````

### The Syntax
```python
newlist = [expression for item in iterable if condition == True]
```
The return value is a new list, leaving the old list unchanged. The condition is like a filter that only accepts the items that evaluate to True.

````python
# Example: Only accept items that are not "apple":
# TODO
newlist = [fruit for fruit in fruits if fruit != "apple"]
newlist
````

### Iterable
The iterable can be any iterable object, like a list, tuple, set etc. You can use conditions.

````python
# Example: Accept only numbers lower than 5:
# TODO
newlist = [x for x in range(10) if x < 5]
newlist
````

### Expression
The expression is the current item in the iteration, but it is also the outcome, which you can manipulate before it ends up like a list item in the new list. The expression can also contain conditions, not like a filter, but as a way to manipulate the outcome.

````python
# Example: Return "orange" instead of "banana"
# TODO
newlist = [fruit if fruit != "banana" else "orange" for fruit in fruits]
newlist
````
