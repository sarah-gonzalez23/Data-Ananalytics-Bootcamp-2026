```python
# Task 6: Function Composition
# Create two functions:
# 1. 'celsius_to_fahrenheit' that converts a temperature from Celsius to Fahrenheit
# 2. 'fahrenheit_to_celsius' that converts a temperature from Fahrenheit to Celsius
# Formula: F = C * 9/5 + 32
# Your code here


# Test Task 6
print("Task 6 Test:")
# Convert some temperatures both ways
# Print the results
print()
```

```python
def celsius_to_fahrenheit(celsius):
    return celsius * 9/5 + 32

def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

print("Task 6 Test:")
print(f"0°C is {celsius_to_fahrenheit(0)}°F")
print(f"100°C is {celsius_to_fahrenheit(100)}°F")

print(f"212°F is {fahrenheit_to_celsius(212)}°C")
print(f"32°F is {fahrenheit_to_celsius(32)}°C")
print()
```

```python
# Task 7: Combining Functions and Conditionals
# Create a function named 'analyze_temperature' that:
# 1. Takes a temperature value and a unit ('C' or 'F')
# 2. Converts the temperature to the other unit
# 3. Returns a string describing if it's "freezing" (below 0°C), 
#    "cold" (0-15°C), "moderate" (15-25°C), or "hot" (above 25°C)
# Your code here


# Test Task 7
print("Task 7 Test:")
# Test with different temperatures in both Celsius and Fahrenheit
# Print the results
print()
```

```python
def celsius_to_fahrenheit(c):
    return c * 9/5 + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

def analyze_temperature(temp, unit):
    if unit.upper() == 'F':
        temp_c = fahrenheit_to_celsius(temp)
        other_temp = temp
        other_unit = 'F'
    else:
        temp_c = temp
        other_temp = celsius_to_fahrenheit(temp)
        other_unit = 'F' if unit.upper() == 'C' else 'C' 


    if temp_c < 0:
        description = "freezing"
    elif 0 <= temp_c <= 15:
        description = "cold"
    elif 15 < temp_c <= 25:
        description = "moderate"
    else:
        description = "hot"

    return f"{temp}°{unit.upper()} is {description} (approx. {other_temp:.1f}°{other_unit})"


print("Task 7 Test:")
print(analyze_temperature(32, 'F'))   
print(analyze_temperature(10, 'C'))   
print(analyze_temperature(72, 'F'))   
print(analyze_temperature(30, 'C'))  
print()
```

```python
# Task 8: Function with Variable Arguments
# Create a function named 'calculate_average' that:
# 1. Accepts any number of numeric arguments
# 2. Returns the average of all numbers provided
# Your code here


# Test Task 8
print("Task 8 Test:")
# Call your calculate_average function with different sets of numbers
# Print the results
print()
```

```python
def calculate_average(*numbers):
    if not numbers:
        return 0
    
    total = sum(numbers)
    count = len(numbers)
    return total / count

print("Task 8 Test:")
print(f"Average of (10, 20, 30): {calculate_average(10, 20, 30)}")
print(f"Average of (5, 15): {calculate_average(5, 15)}")
print(f"Average of no numbers: {calculate_average()}")
print()
```

```python
# Task 9: Advanced Challenge - Student Grade Function
# Create a function named 'calculate_final_grade' that:
# 1. Takes a student's name and any number of assignment scores
# 2. Calculates the average score
# 3. Determines the letter grade based on the average (A: 90+, B: 80-89, C: 70-79, D: 60-69, F: <60)
# 4. Returns a formatted string with the student's name, average score, and letter grade
# Your code here


# Test Task 9
print("Task 9 Test:")
# Test with different student names and scores
# Print the results
print()
```

```python
def calculate_final_grade(student_name, *scores):
    
    if not scores:
        return f"Student {student_name} has no scores recorded."
    
    average_score = sum(scores) / len(scores)
    
    if average_score >= 90:
        letter_grade = 'A'
    elif average_score >= 80:
        letter_grade = 'B'
    elif average_score >= 70:
        letter_grade = 'C'
    elif average_score >= 60:
        letter_grade = 'D'
    else:
        letter_grade = 'F'
    
    return f"Student: {student_name} | Average: {average_score:.1f} | Grade: {letter_grade}"

print("Task 9 Test:")
print(calculate_final_grade("Brittney", 95, 88, 92)) 
print(calculate_final_grade("Anura", 70, 85, 78))    
print(calculate_final_grade("Milton", 100, 98))      
print()
```
